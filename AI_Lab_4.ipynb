```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 5, 7, 8, 9])
y = np.array([0, 6, 4, 6, 7])

plt.subplot(2, 2, 1)
plt.title("One");

plt.plot(x,y)
plt.grid(color = 'pink', linestyle = '--', linewidth = 1)
x = np.array([0, 5, 7, 8, 9])
y = np.array([0, 6, 4, 6, 7])

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.grid(color = 'pink', linestyle = '--', linewidth = 1)
x = np.array([0, 5, 7, 8, 9])
y = np.array([0, 6, 4, 6, 7])

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.grid(color = 'pink', linestyle = '--', linewidth = 1)
x = np.array([0, 2, 3, 5, 6, 8])
y = np.array([0, 3, 2, 4, 3, 5])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.grid(color = 'pink', linestyle = '--', linewidth = 1)
plt.show()
```
